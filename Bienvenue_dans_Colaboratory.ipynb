## Science des données
</div>

Colab vous permet de tirer pleinement parti des bibliothèques populaires Python pour analyser et visualiser des données. La cellule de code ci-dessous utilise <strong>numpy</strong> pour générer des données aléatoires et <strong>matplotlib</strong> pour les visualiser. Pour modifier le code, cliquez simplement sur la cellule.

```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

```
import networkx as nx
import matplotlib.pyplot as plt
import heapq

def bfs(graph, start, goal):
    visited = set()
    queue = [start]
    path = {start: None}

    while queue:
        node = queue.pop(0)
        if node == goal:
            break
        for neighbor in graph[node]:
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append(neighbor)
                path[neighbor] = node

    return reconstruct_path(path, start, goal)

def dfs(graph, start, goal):
    visited = set()
    stack = [start]
    path = {start: None}

    while stack:
        node = stack.pop()
        if node == goal:
            break
        for neighbor in graph[node]:
            if neighbor not in visited:
                visited.add(neighbor)
                stack.append(neighbor)
                path[neighbor] = node

    return reconstruct_path(path, start, goal)

def a_star(graph, start, goal, heuristic):
    open_set = [(0, start)]
    heapq.heapify(open_set)
    came_from = {start: None}
    g_score = {node: float('inf') for node in graph}
    g_score[start] = 0

    while open_set:
        _, current = heapq.heappop(open_set)
        if current == goal:
            break
        for neighbor in graph[current]:
            temp_g_score = g_score[current] + 1
            if temp_g_score < g_score[neighbor]:
                g_score[neighbor] = temp_g_score
                priority = temp_g_score + heuristic[neighbor]
                heapq.heappush(open_set, (priority, neighbor))
                came_from[neighbor] = current

    return reconstruct_path(came_from, start, goal)

def reconstruct_path(came_from, start, goal):
    path = []
    node = goal
    while node is not None:
        path.append(node)
        node = came_from[node]
    return list(reversed(path))

# Définition des nœuds extraits du fichier SVG
nodes = ["A", "1", "2", "3", "4", "5", "10", "6", "7", "8", "9", "11", "12", "13", "14", "15", "17", "16", "18", "20", "19", "21", "B"]

# Définition des arêtes extraites du fichier SVG
edges = [
    ("A", "1"), ("1", "2"), ("2", "3"), ("2", "4"), ("3", "5"), ("4", "10"),
    ("5", "6"), ("5", "7"), ("6", "8"), ("7", "9"), ("10", "11"), ("10", "12"),
    ("12", "13"), ("13", "14"), ("14", "15"), ("14", "17"), ("15", "16"), ("17", "18"),
    ("17", "20"), ("18", "19"), ("20", "21"), ("21", "B")
]

# Définition des positions réajustées pour un affichage vertical
pos = {
    "A": (0, 10), "1": (0, 9), "2": (0, 8), "3": (-1, 7), "4": (1, 7), "5": (-1, 6),
    "10": (1, 6), "6": (-2, 5), "7": (0, 5), "8": (-2, 4), "9": (0, 4), "11": (1, 5),
    "12": (2, 5), "13": (2, 4), "14": (2, 3), "15": (1, 2), "17": (3, 2), "16": (1, 1),
    "18": (3, 1), "20": (4, 1), "19": (3, 0), "21": (4, 0), "B": (4, -1)
}

# Création du graphe avec NetworkX
G = nx.DiGraph()
G.add_nodes_from(nodes)
G.add_edges_from(edges)

# Conversion en dictionnaire d'adjacence
graph = nx.to_dict_of_lists(G)

# Heuristique pour A*
heuristic = {node: abs(pos[node][1] - pos["B"][1]) for node in G.nodes()}

# Test des algorithmes
start, goal = "A", "B"
bfs_path = bfs(graph, start, goal)
dfs_path = dfs(graph, start, goal)
a_star_path = a_star(graph, start, goal, heuristic)

print("BFS Path:", bfs_path)
print("DFS Path:", dfs_path)
print("A* Path:", a_star_path)

# Dessiner le graphe
plt.figure(figsize=(8, 12))
nx.draw(G, pos, with_labels=True, node_color="lightblue", edge_color="gray", node_size=2000, font_size=10, font_weight="bold")
plt.title("Graphe extrait du fichier SVG (Organisation verticale)")

# Sauvegarder l'image au format PNG
plt.savefig("graph.png", format="PNG")

# Afficher le graphe
plt.show()
```
